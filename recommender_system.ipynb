```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import os 
```

```python
movies_data =  pd.read_csv("data_for_movies/tmdb_5000_movies.csv")
credits_data = pd.read_csv("data_for_movies/tmdb_5000_credits.csv")
```

```python
credits_data.head(2)
```

```python
movies_data.shape
```

```python
credits_data.shape
```

```python
movies= movies_data.merge(credits_data, on='title')
# we merged two dataset cause we should focus on one dataset instead of having them seperately 
```

```python
movies.head()
```

```python
# now we need to get rid of the columns that is not necessarily important for our recommendation system
movies.columns
```

```python
# picking up important columns for the recommender system
movies = movies[['movie_id','title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.shape
```

```python
movies.head()
# right now some columns has data in a form of list of dictionaries, so we should do some preprocessing of the data
```

```python
# checking if there is any missing values across the dataset
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)  # we dropped the three null columns 
```

```python
movies.shape
```

```python
movies.duplicated().sum() # no duplicates
```

```python
movies.head(2)
```

```python
movies.iloc[0]['genres'] # it is in a form of str so we cannot iterate it
```

```python
import ast # helps to convert str --> list
def convert(text):
    list = []
    for i in ast.literal_eval(text):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
def convert_cast(text):
    list = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            list.append(i['name'])
        counter += 1
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def fetch_director(text):
    list = []
    counter = 0
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.iloc[0]['overview']
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else x if isinstance(x, list) else [])
movies.head()
```

```python
movies.head()
```

```python
# In columns genres, keywords, cast and crew we can see that there is unneccessary spaces between the words that may confuse the recommendation system
# Like: Sam Worthington in cast and Sam Mendes in crew, that is why we have to get rid of the spaces 


def get_rid_of_spaces(word):
    list = []
    for i in word:
        list.append(i.replace(" ", "")) 
    return list 
```

```python
movies['cast'] = movies['cast'].apply(get_rid_of_spaces)
movies['crew'] = movies['crew'].apply(get_rid_of_spaces)
movies['genres'] = movies['genres'].apply(get_rid_of_spaces)
movies['keywords'] = movies['keywords'].apply(get_rid_of_spaces)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head()
```

```python
# now  we reconvert list --> str
new_df = new_df.copy()
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
import nltk
from nltk.stem import PorterStemmer 
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df [new_df['title'] == 'Spider-Man'].index[0]
```

```python
def recommend(movie):
    index = new_df [new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
recommend('Avatar')
```

```python
recommend('The Dark Knight Rises')
```
